```python
import pandas as pd
import numpy  as np

import sys

import requests
from bs4 import BeautifulSoup 
```

```python
html = requests.get('https://www.tripadvisor.cl/Restaurant_Review-g294305-d10503885-Reviews-Krossbar-Santiago_Santiago_Metropolitan_Region.html')
html.status_code

```

```python
bsobj = BeautifulSoup(html.content, 'lxml')
# Imprime todo el HTML que queremos scrapear
# print(bsobj)
```

```python
# Listado de Links
links = []

# Appending el primer link
links.append('https://www.tripadvisor.cl/Restaurant_Review-g294305-d10503885-Reviews-Krossbar-Santiago_Santiago_Metropolitan_Region.html')

# Variable para navegar p√°ginas de comentarios
x = 10


# 59 p√°ginas de comentarios en total
# Se ocupa 580 porque la primera pagina tiene comentarios pero ningun
# -orXX para contar

while x <= 580:
    a = '/Restaurant_Review-g294305-d10503885-Reviews-Krossbar-Santiago_Santiago_Metropolitan_Region.html'
    a = 'https://www.tripadvisor.cl' + a    
    
    # -or{} cambia en variables de 10
    
    c = f'-or{x}'
    a = a[:(a.find('Reviews') + 7)] + c + a[(a.find('Reviews') + 7):]
    
    print(a)
    
    links.append(a)
    
    x += 10
    
    
```

```python
# Creating our dataframe to store info
df = pd.DataFrame({'titulo':[], 'text':[], 'date':[] })
df
```

```python
y = 0

html = requests.get(links[y])
soup = BeautifulSoup(html.content, 'lxml')



# Corriendo el loop hasta que se termine la lista.
while y < 59:
    reviews = soup.find_all('div', class_ = 'review-container')
    for review in reviews:
        # Encontrando la Fecha de Cada Review
        fecha_review = review.find('div', class_ = 'prw_rup prw_reviews_stay_date_hsx')\
        .text.replace("Fecha de la visita: ", "")

        # El Titulo
        titulo_review = review.find('span', class_ = 'noQuotes').text

        # El Comentario
        text_review = review.find('div', class_ = 'entry' ).text
                
        # Imprimiendo los resultados
        print(f'''{fecha_review} {titulo_review}
        {text_review}

        ''')
            
            
        # Appending los resultados a nuestro df
        new_row = {'titulo':titulo_review,'text':text_review,'date':fecha_review}
        df = df.append(new_row, ignore_index = True)
            
    # Yendo al siguente link
    html = requests.get(links[y])
    soup = BeautifulSoup(html.content, 'lxml')
    y += 1


            

            
# Imprimiendo nuestro resultado.
df
  
```
